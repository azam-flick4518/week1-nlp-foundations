```python
# What is a token?
text = "I love building AI systems."

# Split by whitespace — naive tokenization
words = text.split()
print(f"Words: {words}")
print(f"Word count: {len(words)}")
```

```python
# Real tokenizer — how transformers actually see text
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
tokens = tokenizer.tokenize(text)
ids = tokenizer.encode(text)

print(f"Tokens: {tokens}")
print(f"Token IDs: {ids}")
print(f"Token count: {len(tokens)}")
```

```python
# Decode the special tokens
print(tokenizer.decode([101]))
print(tokenizer.decode([102]))
print(f"\nVocabulary size: {tokenizer.vocab_size}")
```

```python
# What happens with words NOT in the vocabulary?
words = ["unbelievable", "ChatGPT", "azam", "transformerized"]

for word in words:
    tokens = tokenizer.tokenize(word)
    print(f"{word:20} → {tokens}")
```

```python
# Cost and context implications
import pandas as pd

samples = [
    "I love building AI systems.",
    "ChatGPT is an unbelievable transformerized model.",
    "azam charaniya is learning NLP in Sydney.",
    "The quick brown fox jumps over the lazy dog."
]

data = []
for s in samples:
    tokens = tokenizer.tokenize(s)
    data.append({
        "text": s,
        "words": len(s.split()),
        "tokens": len(tokens),
        "ratio": round(len(tokens) / len(s.split()), 2)
    })

df = pd.DataFrame(data)
print(df.to_string(index=False))
```

A token is a numerical ID representing a piece of text (like a syllable or character), while a word is a human unit that often requires multiple tokens to represent.
API costs are calculated per token, so text that breaks into more pieces is more expensive to process, even if the word count is the same.
Since a model's context window is a fixed size (e.g., 128k tokens), a higher token-per-word ratio fills that space faster, leaving less room for the rest of your data.
